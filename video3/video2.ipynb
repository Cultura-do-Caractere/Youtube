```python
from pycoingecko import CoinGeckoAPI
cg = CoinGeckoAPI()
coin_price = cg.get_price(ids="bitcoin", vs_currencies='eur')
btc_value = coin_price.get("bitcoin").get("eur")
btc_value
```

```python
#pd.set_option('display.float_format', lambda x: '%.10f' % x)
btc_transactions = crypto_transactions[crypto_transactions["Transaction Description"] == "Buy BTC"]
btc_transactions
```

```python
btc_amount = btc_transactions["Amount"].sum()
btc_amount
```

```python
money_invested = btc_transactions["Native Amount"].sum()
money_invested
```

```python
current_amount = btc_amount * btc_value
current_amount
```
